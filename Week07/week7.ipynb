```python
import numpy as np
import pandas as pd
from scipy.integrate import quad as integrate
```

```python
def GLCG(x0):
    a,b,m = 625,6571,31104
    while True:
        x0 = (a*x0+b)%m
        yield x0/m

glcg = GLCG(69)

def U(a,b):
    return a+(b-a)*next(glcg)
```

```python
E = 2
S = integrate(lambda x: np.exp(2*x**0.5), 0,1)[0] - E**2
S = S**0.5
```

```python
results = {}
for N in (10**2//2,10**2,10**3//2,10**3,10**4//2,10**4,10**5//2,10**5):
    print("N =",N)
    
    I1 = list(
            np.exp(U(0,1)**0.5) 
        for i in range(N))
    
    I2 = list(
            (np.exp(U(0,1)**0.5)+ np.exp((1-U(0,1))**0.5))/2
        for i in range(N))
        
    E1 = sum(I1)/N
    E2 = sum(I2)/N
    
    Sn1 = sum( (x-E1)**2 for x in I1 )/(N-1)
    Sn1 = Sn1**0.5
    Sn2 = sum( (x-E2)**2 for x in I2 )/(N-1)
    Sn2 = Sn2**0.5
    
    inter1 = (E1 - 1.96 * Sn1/N**0.5, E1 + 1.96 * Sn1/N**0.5)
    inter2 = (E2 - 1.96 * Sn2/N**0.5, E2 + 1.96 * Sn2/N**0.5)
    interval_true = (E - 1.96 * S/N**0.5, E + 1.96 * S/N**0.5)

    inter1 = tuple(map(lambda x: round(x,4), inter1))
    inter2 = tuple(map(lambda x: round(x,4), inter2))
    interval_true = tuple(map(lambda x: round(x,4), interval_true))
    
    results[N] = [E1, E2, inter1, inter2, interval_true, (inter1[1]-inter1[0])/(inter2[1]-inter2[0])]
    
    
results = pd.DataFrame(results).T
results.columns = ["E1","E2", "Interval 1", "Interval 2", "True interval", "Ratio |I1| / |I2|"]


results
```
