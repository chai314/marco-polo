# Week 1

A general-linear-congruence generator class that I will use as RNG

```python
import numpy as np
import matplotlib.pyplot as plt
class GLCG:
    def __init__(self,a,b,m):
        self.a=a; self.b=b; self.m=m

    def seed(self,x):
        # Set seed
        self.x0=x
        print("Seed set to",self.x0)

    def summary(self):
        print(f'''Running parameters: a={self.a}
                    b={self.b}
                    m={self.m}
              ''')
        
    def gen(self, mode="x"):
        # Generator output
        divisor = 1 if mode=="x" else self.m
        seen = set()
        x = self.x0
        while x not in seen:
            seen|={x}
            yield x/divisor
            x = (self.a*x+self.b)%self.m
```

```python
rng = GLCG(6,0,11)
rng.seed(0)
rng.summary()
```

## Problem 1.

```python
for seed in range(11):
    rng.seed(seed)
    L = [*rng.gen()]
    print(L)
    print(len(L),"distinct\n")
```

Observation:

Seed = 0 is trivial and not useful 

In the nontrivial case, numbers follow the same cycle before they loop. 

The cycle is <5.0, 8.0, 4.0, 2.0, 1.0, 6.0, 3.0, 7.0, 9.0, 10.0>

```python
#Part 2
rng = GLCG(3,0,11)
rng.seed(0)
rng.summary()

for seed in range(11):
    rng.seed(seed)
    L = [*rng.gen()]
    print(L)
    print(len(L),"distinct\n")
    
```

Observation:


The period is 5 here which is less than 10.

Depending on seed, the cycle can be [4.0, 1.0, 3.0, 9.0, 5.0] or [10.0, 8.0, 2.0, 6.0, 7.0].

### Conclusion:
We can see that the generator in part 1 is the better one, as it runs for a full period giving us more random values than the second generator.

## Problem 2

```python
a = 1597
b = 51749
m = 244944

seeds = (0,1,122472, 23, 244944)
for x in seeds:
    print(f"Using seed {x}")
    plt.title(f'Seed = {x}')
    freq = np.zeros(20)
    for _ in range(10000):
        x = (a*x + b)%m / m
        freq[int(x/0.05)] += 1
    X_axis = [i*0.05 for i in range(0, 20)]
    print(f"Frequency of each range: {freq}")
    plt.bar(X_axis, freq, width = 0.02, align = 'edge')
    plt.xlabel('Range')
    plt.ylabel('Frequency')
    plt.show()
```

### Conclusion:
For different seeds, we find that the numbers generated are roughly equal in all the buckets. Regardless of the seed value, we find the numbers to be particularly spaced out and sufficiently random, indicating that the choice of a, b, and m is a good one. 
The seed values tried were, 0, 1, 122472, 23, 244944
